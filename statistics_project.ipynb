```python
import pandas as pd

df_test = pd.read_csv('test.csv')
df_train = pd.read_csv('train.csv')
```

```python
df_train.sort_values('customer_id', inplace=True)
df_train
```

```python
df_test.sort_values('customer_id', inplace=True)
df_test
```

```python
df_test_sub = df_test.drop(columns=['customer_id','signup_channel','country',])
df_test_sub['premium_user'] = df_test_sub['premium_user'].replace({'Y': 1, 'YES': 1, 'Yes': 1, 'yes': 1, 'y': 1, 'N': 0, 'NO': 0, 'No': 0, 'no': 0, 'n': 0,' YES ':1, ' NO ':0})
df_test_sub['premium_user'] = df_test_sub['premium_user'].astype(int)
df_test_sub
```

```python
df_test_sub.info()
```

```python
df_train_sub = df_train.drop(columns=['customer_id','signup_channel','country',])
df_train_sub['premium_user'] = df_train_sub['premium_user'].replace({'Y': 1, 'YES': 1, 'Yes': 1, 'yes': 1, 'y': 1, 'N': 0, 'NO': 0, 'No': 0, 'no': 0, 'n': 0,' YES ':1, ' NO ':0})
df_train_sub['premium_user'] = df_train_sub['premium_user'].astype(int)
df_train_sub
```

```python

X_train, y_train = df_train_sub.to_numpy()[:, :-1], df_train_sub.to_numpy()[:, -1]
y_train

```

```python
from sklearn.linear_model import LogisticRegression

# Fix: instantiate LogisticRegression first, then call fit()
lr = LogisticRegression().fit(X_train, y_train)
lr
```
