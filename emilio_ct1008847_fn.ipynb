```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, median_absolute_error,  r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.cluster import KMeans


diabetes = load_diabetes()
```

Your tasks include:

Prepare training and testing sets with an 80-20 split ratio. This helps in machine learning, with 80% of the data used for training and 20% for testing.

Conduct regression using the Random Forest ensemble technique.

You are required to set n_estimators=(the last two digits of your student ID multiplied by 10). If the last two digits are 00, use 10 as the n_estimators

For example, if the last two digits of the student ID are 34, then n_estimators=340
Compute the R-squared value to evaluate your model and at least two other metrics for the Random Forest regression model.

Conduct regression using the XGBoost ensemble technique. Use the same n_estimators value as used in Question 1.2.

```python
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
```

```python
df.describe().round(2)
```

```python
df.info()
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_df
```

```python
test_df
```

```python
# Replace 'XX' with the last two digits of your student ID
last_two_digits = 47
X_train = train_df.drop('target', axis=1)
y_train = train_df['target']
X_test = test_df.drop('target', axis=1)
y_test = test_df['target']


n_estimators = last_two_digits * 10 if last_two_digits != 0 else 10
eg_rf = RandomForestRegressor(n_estimators=n_estimators, random_state=42)
eg_rf.fit(X_train, y_train)

# Predict on test set
y_pred = eg_rf.predict(X_test)

# Evaluate performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Random Forest MSE: {mse:.2f}")
print(f"Random Forest R^2: {r2:.2f}")
\
# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")

# Median Absolute Error
medae = median_absolute_error(y_test, y_pred)
print(f"Median Absolute Error: {medae:.2f}")
```

```python
# Train XGBoost regressor with the same n_estimators
eg_xgb_reg = XGBRegressor(n_estimators=n_estimators, random_state=42)
eg_xgb_reg.fit(X_train, y_train)

# Predict on test set
y_pred_xgb = eg_xgb_reg.predict(X_test)

# Evaluate performance

mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print(f"XGBoost MSE: {mse_xgb:.2f}")
print(f"XGBoost R^2: {r2_xgb:.2f}")
# Mean Absolute Error (MAE)
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
print(f"XGBoost Mean Absolute Error: {mae_xgb:.2f}")

# Median Absolute Error
medae_xgb = median_absolute_error(y_test, y_pred_xgb)
print(f"XGBoost Median Absolute Error: {medae_xgb:.2f}")
```
